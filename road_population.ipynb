```python
#행정구역 column에서 읍 면 동 셋 중 하나로 끝나는 단어만 출력
import re
import numpy as np
population = pd.read_csv('data/제주시_주민등록인구및세대현황 final.csv', encoding='utf-8', sep=';')
def add_space_to_address_parts(행정구역):
    return re.sub(r'(읍|면|동)\(', r'\1 (', 행정구역)

population['행정구역'] = population['행정구역'].apply(add_space_to_address_parts)

def filter_address_parts(population):
    parts = population.split()
    filtered_parts = [part for part in parts if part.endswith(('읍', '면', '동'))]
    return ' '.join(filtered_parts)

population['filtered_address'] = population['행정구역'].apply(filter_address_parts)


#nodes count column 만들어서 추가 행정동 기준으로 분류되어 있으므로 각 다음과 같은 법정동을 포함한다. 제주시 봉개동: 봉개동, 용강동, 회천동/ 노형동: 노형동, 해안/ 외도동: 내도동, 도평동
node_counts = [0, 194, 304, 178, 207, 105, 0, 0, 18, 19, 19, 98, 11, 25, 29, 35, 40, 94, 37, 58, 87, 60, 106, 225, 55, 23, 37]
population['nodes count'] = node_counts


#node 당 averge population을 구한다
population['2023년10월_총인구수'] = population['2023년10월_총인구수'].astype(str)
population['2023년10월_총인구수'] = population['2023년10월_총인구수'].str.replace('.', '').astype(int)
population['average population'] = population['2023년10월_총인구수'] / population['nodes count']
population['average population'] = population['average population'].replace([np.inf, -np.inf], np.nan)

population.to_csv('data/제주시_주민등록인구및세대현황 final.csv', encoding='utf-8', sep=';')
```
