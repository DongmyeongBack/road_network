```python
import pandas as pd
import numpy as np
import networkx as nx

import matplotlib.pyplot as plt
# 읍 면 동이 행정동으로 구별된 것을 지도에 그리기
```

```python
nodes = pd.read_csv('data/Jeju_nodes.csv', encoding='utf-8',sep=',')
links = pd.read_csv('data/Jeju_links.csv', encoding='cp949')
links
nodes = nodes[['Id','NODE_NAME','latitude','longitude','filtered_address','average population']]
links = links[['Source','Target']]
```

```python
source_in = links['Source'].apply(lambda x : x in list(nodes['Id'])) # check Sources are in jeju_id
target_in = links['Target'].apply(lambda x : x in list(nodes['Id'])) # check Targets are in jeju_id
# source_in and target_in are boolean type pandas.Series which contains True or False
Jeju_links = links[source_in & target_in] # contain if both target and source are contained in jeju_id
```

```python
G = nx.Graph()
# R is the Earth's radius
R = 6371e3

for idx,row in nodes.iterrows():
    # add node to Graph G
    G.add_node(row['Id'],Label=row['NODE_NAME'],latitude=row['latitude'], longitude=row['longitude'])

for idx,row in Jeju_links.iterrows():
    ## Calculate the distance between Source and Target Nodes
    lon1 = float(nodes[nodes['Id'] == row['Source']]['longitude'] * np.pi/180)
    lat1 = float(nodes[nodes['Id'] == row['Source']]['latitude'] * np.pi/180)
    lon2 = float(nodes[nodes['Id'] == row['Target']]['longitude'] * np.pi/180)
    lat2 = float(nodes[nodes['Id'] == row['Target']]['latitude'] * np.pi/180)
    d_lat = lat2 - lat1
    d_lon = lon2 - lon1
    a = np.sin(d_lat/2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(d_lon/2) ** 2
    c = 2 * np.arctan2(a**0.5, (1-a) ** 0.5)
    d = R * c
    
    # Link attribute : 'Source', 'Target' and weight = 'Length between them'
    G.add_edge(row['Source'],row['Target'],weight = d)
    
# make a dictionary for node position
pos = {}
for node in G.nodes(data=True):
    pos[node[0]] = [node[1]['longitude'], node[1]['latitude']]
print(pos)
```

```python
import pandas as pd
import networkx as nx
from haversine import haversine
import matplotlib.pyplot as plt


nodes = pd.read_csv('data/Jeju_nodes.csv', encoding='utf-8')
links = pd.read_csv('data/Jeju_links.csv', encoding='cp949')
car = pd.read_csv('data/Jeju_car.csv', encoding='cp949')


nodes = nodes[['Id', 'NODE_NAME', 'latitude', 'longitude', 'average population']]
links = links[['Source', 'Target', '시간']]
source_in = links['Source'].apply(lambda x: x in list(nodes['Id']))
target_in = links['Target'].apply(lambda x: x in list(nodes['Id']))
jeju_links = links[source_in & target_in]

```

```python
G = nx.Graph()

for idx, row in nodes.iterrows():
    G.add_node(row['Id'], Label=row['NODE_NAME'], latitude=row['latitude'], longitude=row['longitude'], avg_pop=row['average population'])

for idx, row in jeju_links.iterrows():
    G.add_edge(row['Source'], row['Target'], weight=row['시간'])

def connectRes(graph, df, nodes_df):
    nodes_latitude = nx.get_node_attributes(graph, 'latitude')
    nodes_longitude = nx.get_node_attributes(graph, 'longitude')
    nodes_avg_pop = nx.get_node_attributes(graph, 'avg_pop')
    for idx, row in df.iterrows():
        distances = {}
        for key, val in nodes_latitude.items():
            nodesPos = (val, nodes_longitude[key])
            curPos = (row['latitude'], row['longitude'])
            distances[key] = haversine(nodesPos, curPos, unit='km')
        nearest_node = min(distances, key=distances.get)
        avg_pop = nodes_avg_pop[nearest_node]
        graph.add_node(idx, Label=idx, latitude=row['latitude'], longitude=row['longitude'], avg_pop=avg_pop)
        w = distances[nearest_node] * 1000 / 666
        graph.add_edge(nearest_node, idx, weight=w)
    return graph

G = connectRes(G, car, nodes)
```

```python

pos = {node[0]: [node[1]['longitude'], node[1]['latitude']] for node in G.nodes(data=True)}

def propagation(G, start_node, w=5):
    shortest_paths = nx.single_source_dijkstra_path_length(G, start_node, cutoff=w, weight='weight')
    return shortest_paths

def mergeDict(dict1, dict2):
    result_dict = {}
    for key in set(dict1.keys()).intersection(dict2.keys()):
        result_dict[key] = min(dict1[key], dict2[key])
    result_dict.update({key: dict1[key] for key in set(dict1.keys()) - set(dict2.keys())})
    result_dict.update({key: dict2[key] for key in set(dict2.keys()) - set(dict1.keys())})
    return result_dict

```

```python
serviceArea = {}
for idx, row in car.iterrows():
    tmp = propagation(G, idx, w=5)
    serviceArea[idx] = list(tmp.keys())  # Ensure each service area maps to a list of nodes


```

```python
import pandas as pd
import numpy as np
import networkx as nx

import matplotlib.pyplot as plt
# 읍 면 동이 행정동으로 구별된 것을 지도에 그리기
nodes = pd.read_csv('data/Jeju_nodes.csv', encoding='utf-8',sep=',')
links = pd.read_csv('data/Jeju_links.csv', encoding='cp949')
links
nodes = nodes[['Id','NODE_NAME','latitude','longitude','average population']]
links = links[['Source','Target']]
source_in = links['Source'].apply(lambda x : x in list(nodes['Id'])) # check Sources are in jeju_id
target_in = links['Target'].apply(lambda x : x in list(nodes['Id'])) # check Targets are in jeju_id
# source_in and target_in are boolean type pandas.Series which contains True or False
Jeju_links = links[source_in & target_in] # contain if both target and source are contained in jeju_id
G = nx.Graph()
# R is the Earth's radius
R = 6371e3
```

```python
for idx,row in nodes.iterrows():
    # add node to Graph G
    G.add_node(row['Id'],Label=row['NODE_NAME'],latitude=row['latitude'], longitude=row['longitude'])

for idx,row in Jeju_links.iterrows():
    ## Calculate the distance between Source and Target Nodes
    lon1 = float(nodes[nodes['Id'] == row['Source']]['longitude'] * np.pi/180)
    lat1 = float(nodes[nodes['Id'] == row['Source']]['latitude'] * np.pi/180)
    lon2 = float(nodes[nodes['Id'] == row['Target']]['longitude'] * np.pi/180)
    lat2 = float(nodes[nodes['Id'] == row['Target']]['latitude'] * np.pi/180)
    d_lat = lat2 - lat1
    d_lon = lon2 - lon1
    a = np.sin(d_lat/2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(d_lon/2) ** 2
    c = 2 * np.arctan2(a**0.5, (1-a) ** 0.5)
    d = R * c
    
    # Link attribute : 'Source', 'Target' and weight = 'Length between them'
    G.add_edge(row['Source'],row['Target'],weight = d)
```

```python
# make a dictionary for node position
pos = {}
for node in G.nodes(data=True):
    pos[node[0]] = [node[1]['longitude'], node[1]['latitude']]
print(pos)
from matplotlib.colors import ListedColormap

```

```python
# 노드별 인구수 기준으로 색상 맵 생성
unique_populations = nodes['average population'].unique()
max_population = nodes['average population'].max()
color_map = plt.cm.Blues  

# 인구수에 따른 색상 할당
population_color = {population: color_map(population / max_population) for population in unique_populations}


def get_node_attributes(node):
    avg_population = nodes[nodes['Id'] == node]['average population'].values[0]
    color = population_color[avg_population]
    size = 10 
    if avg_population == 0:
        color = 'white'  
    return color, size


node_colors = [get_node_attributes(node)[0] for node in G.nodes()]
node_sizes = [get_node_attributes(node)[1] for node in G.nodes()]


def draw_graph(G, pos, node_colors, node_sizes):
    plt.figure(figsize=(12, 7))
    plt.axis('off')
    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=node_sizes)
    nx.draw_networkx_edges(G, pos, edge_color='grey')
    plt.show()


draw_graph(G, pos, node_colors, node_sizes)
```

```python
import matplotlib.colors as mcolors

def get_edge_color(source, target, node_colors_dict):
    # 두 노드의 색상 평균 계산 후 링크에 넣기
    color_source = mcolors.to_rgb(node_colors_dict[source])
    color_target = mcolors.to_rgb(node_colors_dict[target])
    avg_color = [(cs + ct) / 2 for cs, ct in zip(color_source, color_target)]
    return mcolors.to_hex(avg_color)

node_colors_dict = {node: get_node_attributes(node)[0] for node in G.nodes()}


edge_colors = [get_edge_color(u, v, node_colors_dict) for u, v in G.edges()]


def draw_graph(G, pos, node_colors, node_sizes, edge_colors):
    plt.figure(figsize=(12, 7))
    plt.axis('off')
    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=node_sizes)
    nx.draw_networkx_edges(G, pos, edge_color=edge_colors)
    plt.show()


draw_graph(G, pos, node_colors, node_sizes, edge_colors)
```
