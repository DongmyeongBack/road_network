```python
import pandas as pd
file_path = 'Jeju_traffic.csv'

df = pd.read_csv(file_path)
df.head()
```

```python
column_name=['순번', '생성일', '생성시분', '링크ID', '도로관리기관','통행속도','통행시간']

data = df.values.tolist()
new_df = pd.DataFrame(data, columns=column_name)
new_df
```

```python
idx = new_df.groupby('링크ID')['통행속도'].idxmax()


max_speed_df = new_df.loc[idx]

print(max_speed_df)
```

```python
max_speed_df.to_csv('final.csv', index=False)
```

```python
final_df = pd.read_csv('final.csv')

Jeju_links_df = pd.read_csv('Project/data/Jeju_links.csv')


merged_df = pd.merge(final_df,Jeju_links_df, left_on='링크ID', right_on='LINK_ID', how='inner')


merged_df.drop('LINK_ID', axis=1, inplace=True)


merged_df.to_csv('final.csv', index=False)
print(merged_df)
```
